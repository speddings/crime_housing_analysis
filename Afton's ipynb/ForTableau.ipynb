# Clean Fire Data

```python
# read in fire data file
ca_fire_df = pd.read_csv("ca_daily_fire_2000_2021.csv")
ca_fire_df.head()
```

```python
# drop any null values
ca_fire_df.dropna()
```

```python
# change lables 
# df = df.rename(columns={'oldName1': 'newName1', 'oldName2': 'newName2'})
ca_fire_df.rename(columns={'longitude':'lon',
                          'latitude':'lat'}, inplace=True)
ca_fire_df.head()
```

```python
# drop all ROWS not 2019 (retain all 2019 rows)
clean_ca_fire_df = ca_fire_df[ca_fire_df['year']==2019]
clean_ca_fire_df.head()
```

```python
lat_lngs = zip(ca_fire_df['lat'],ca_fire_df['lon'])
# print(len(list(lat_lngs)))
cities = []

# Identify the nearest city for each latitude and longitude combination.
for lat_lng in lat_lngs:
    zipcodes = engine.by_coordinates(lat_lng[0], lat_lng[1], radius=30)

#     if zipcodes[0].state == 'CA':
    try:
        cities.append(zipcodes[0].major_city)
        
    except IndexError:
        cities.append('')
        print('exception')
        
#     print(zipcodes)    
    
len(cities)
```

```python
ca_fire_df['cities'] = pd.DataFrame({'cities': cities})
ca_fire_df
```

```python
# drop all ROWS not 2019 (retain all 2019 rows)
clean_ca_fire_df = ca_fire_df[ca_fire_df['year']==2019]
clean_ca_fire_df.head()
```

```python
clean_ca_fire_df = clean_ca_fire_df.dropna()
clean_ca_fire_df
```
