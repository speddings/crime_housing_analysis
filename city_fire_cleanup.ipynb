```python
# Import modules.
import pandas as pd
import numpy as np
import gmaps
import citipy
from uszipcode import SearchEngine

# Import the API key.
from citipy import citipy
engine = SearchEngine()
```

```python
# read in fire data file
ca_fire_df = pd.read_csv("../Resources/clean_fire.csv")
ca_fire_df
```

```python
ca_fire_df['lat']
```

```python
lat_lngs = zip(ca_fire_df['lat'],ca_fire_df['lon'])
# print(len(list(lat_lngs)))
cities = []

# Identify the nearest city for each latitude and longitude combination.
for lat_lng in lat_lngs:
    zipcodes = engine.by_coordinates(lat_lng[0], lat_lng[1], radius=30)

#     if zipcodes[0].state == 'CA':
    try:
        cities.append(zipcodes[0].major_city)
        
    except IndexError:
        cities.append('')
        print('exception')
        
#     print(zipcodes)    
    
len(cities)
```

```python
ca_fire_df['cities'] = pd.DataFrame({'cities': cities})
ca_fire_df
```

```python
ca_fire_df = ca_fire_df.dropna()
ca_fire_df
```

```python
ca_fire_df.to_csv("../Resources/clean_fire_cities.csv")
```
